```python
df[df["sex"].isna()]
```

```python
df_clean1 = df.dropna()
print(df_clean1)
print(df_clean1.shape,df.shape)
```

```python
df_clean2 = df.copy()
df_clean2["body_mass_g"] = df["body_mass_g"].fillna(df["body_mass_g"].median())
df_clean2["body_mass_g"]
```

Commentaire: on remplace les valeurs manquantes NaN par la médiane  de la colonne au lieu de la moyenne. Cela rend le résultat plus robuste pour les analyses ultérieures parce que la médiane est moins sensible aux valeurs extrêmes (aberrantes) que la moyenne.

# Block 3:  MERGE et JOIN

```python
# on va créer un autre table: nom scientifique des composition des espèces
especes = pd.DataFrame({
    "species" : ["Adelie","Gentoo","Chinstrap"],
    "nom_scientifique" : ["Pygoscelis adeliae", "Pygoscelis papua", "Pygoscelis antarcticus"]
})

especes
```

```python

duckdb.sql("select * from df join especes on df.species=especes.species")
```

```python


especes_nan = pd.DataFrame({
    "species" : ["Adelie", "Gentoo", "Chinstrap"],
    "nom_scientifique" : ["Pygoscelis adeliae", "Pygoscelis papua", np.nan ] # để 1 dòng có giá trị NaN => dùng np.nan hoặc None
})
especes_nan
```

```python

df_merge1 = df.merge(especes_nan, on ="species", how="left")  # Pandas
print(df_merge1.isna().sum())
df_merge1[df_merge1["species"]=="Chinstrap" ]
```

```python
# SQL
duckdb.sql("select count(*) as nb_null from df left join especes1 on df.species = especes1.species where especes1.nom_scientifique is null")
```

```python

duckdb.sql("select * from df left join especes1 e1 on df.species=e1.species where e1.nom_scientifique is null")
```

```python

df_merge = df.merge(especes,on="species")
df_merge.info()
```

```python
df_merge
```

```python
especes_sans_chinstrap = pd.DataFrame({
    "species" : ["Adelie", "Gentoo"],
    "nom_scientifique" : ["Pygoscelis adeliae", "Pygoscelis papua"]
})
merge = df.merge(especes_sans_chinstrap, on="species")
mergeleft = df.merge(especes_sans_chinstrap, on="species",how="left")
print(merge.shape, mergeleft.shape)
```

 Commentaire: à partir du résultat, le nombre de lignes de la table obtenue par inner join est 276, alors que le nombre de lignes de la table obtenu par left join est 344. Cela s'explique par la manière dont la jointure fonctionne. L'inner join supprime automatiquement les lignes de la table de gauche qui n'ont pas de correspondance dans la table de droite. En revanche, left join conserve toutes les lignes de table de gauche, même sans correspondance à droite. Les colonnes manquantes sont alors remplies par NaN.
