# Partie 3: Analyse - Comparaison des trois espèces

```python
# Comparer les trois espèces
# 1: la masse corporelle moyenne
x1 = df[df["species"]=="Adelie"]["body_mass_g"].mean()
x2 = df[df["species"]=="Gentoo"]["body_mass_g"].mean()
x3 = df[df["species"]=="Chinstrap"]["body_mass_g"].mean()
print(x1,x2,x3)
```

```python
# 2: la longueur des nageoires moyenne
y1 = df[df["species"]=="Adelie"]["flipper_length_mm"].mean()
y2 = df[df["species"]=="Gentoo"]["flipper_length_mm"].mean()
y3 = df[df["species"]=="Chinstrap"]["flipper_length_mm"].mean()
print(y1,y2,y3)
```

## Commentaire:
On compare d'abord la masse corporelle moyenne des espèces. L'espèce Gentoo a une masse nettement supérieure à celle des deux autres, soit environ 5076 g, contre environ 3700 g pour les Adelie et 3733 g pour les Chinstrap.
Ensuite, on compare la longueur des nageoires. La même tendance se confirme : les Gentoo ont les nageoires les plus longues, environ 217 mm, tandis que les Adelie et les Chinstrap ont une longueur à peu près équivalente, environ 190 mm et 196 mm.
En résumé, les Gentoo se distinguent nettement des autres par leur grande taille, aussi bien en masse corporelle qu'en longueur des nageoires, alors que les Chinstrap sont légèrement plus grands que les Adelie.
